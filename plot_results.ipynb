```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
def read_excel(file_path):
    df = pd.read_excel(file_path, index_col=0)  
    return df

def plot_episodereward(df):
    x = df["Episode"]
    y = df["Reward Sum"]
    s = df["No of steps"]
    _, b = plt.subplots(1, 2, figsize=(20,10))
    b[0].scatter(x, y)
    b[0].set_xlabel("No of Episode")
    b[0].set_ylabel("Reward Sum")
    b[1].scatter(x, s)
    b[1].set_xlabel("No of Episode")
    b[1].set_ylabel("No of steps")

def plot_losses(df):
    x = df["Episode"]
    legend=[]
    for k in df.columns:
        if "loss" in k:
            plt.scatter(x, df[k])
            legend.append(k)
    plt.legend(legend, loc ="lower right")
    plt.xlabel("No of Episode")
    plt.ylabel("Final loss")
```

```python
dir_path = "/home/carolina/Anansi/RL/weights_NoCheating_v8.0"
file_name = "results"
file_path = "{}/{}.xlsx".format(dir_path, file_name)
df = read_excel(file_path)
```
